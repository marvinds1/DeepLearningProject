Membuat fungsi untuk melakukan preprocessing dan ekstraksi ciri TF-ID pada data teks tanpa menggunakan library dan menggunakan bahasa python melalui google colab.

```
dok1 = "sistem kenal pola titik berat kembang model basis machine learn kenal pola"
dok2 = "tantang kenal pola intern-class variable kondisi objek pola rupa objek ada kelas beda"
dok3 = "normal nilai fitur latih beri akurasi baik sistem kenal pola banding laku normal"
dok4 = "sistem kenal pola kembang laku kasus tentu"
dok5 = "tiga data data teks data memiliki volume banyak struktur proses analisis basis kenal pola mudah laku"
```

```
katad1 = dok1.split()
katad2 = dok2.split()
katad3 = dok3.split()
katad4 = dok4.split()
katad5 = dok5.split()
katad1
```

```
katad12 = set(katad1).union(set(katad2)).union(set(katad3)).union(set(katad4)).union(set(katad5))
```

```
katad12
```

```
tf1 = dict.fromkeys(katad12, 0) 
tf2 = dict.fromkeys(katad12, 0)
tf1
```

```
for word in katad1:
    tf1[word]+=1
    
for word in katad2:
    tf2[word]+=1

tf1
```

```
tf2
```

```
def TermFrequency(tfd, katad):
    temp = {}
    temp2 = len(katad)
    for word, count in tfd.items():
        temp[word] = count/float(temp2)
    return temp

tfd1 = TermFrequency(tf1, katad1)
tfd2 = TermFrequency(tf2, katad2)
tfd1
```

```
tfd2
```

```
def IDF(dList):
    import math
    temp = {}
    temp2 = len(dList)
    
    temp = dict.fromkeys(dList[0].keys(), 0)
    for doc in dList:
        for word, val in doc.items():
            if val > 0:
                temp[word] += 1
    
    for word, val in temp.items():
        temp[word] = 1 + math.log10(temp2 / float(val))
        
    return temp

idf12 = IDF([tf1, tf2])
idf12
```

```
def TFIDF(tfd, idf12):
    temp = {}
    for word, val in tfd.items():
        temp[word] = val*idf12[word]
    return temp

tfidfd1 = TFIDF(tfd1, idf12)
tfidfd2 = TFIDF(tfd2, idf12)
tfidfd1
```
