Import dan ambil semua library/alat yang kita akan pakai

```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
```

Ambil dataset yang ditelah disimpan di dalam drive

```python
habermandata = pd.read_csv('/content/gdrive/MyDrive/ColabNotebooks/haberman.data',header=None)
habermandata.head()
```

Split the dataset dengan tes_size = 0.2

```python
X = habermandata.iloc[:, [0, 1, 2]].values
y = habermandata.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

Scaling dataset

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## **Train Dataset dengan Algoritma KKN**

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_test
```

```python
y_pred
```

```python
MatrixC = confusion_matrix(y_test, y_pred)
ReportC = classification_report(y_test,y_pred)
ScoreA = accuracy_score(y_test,y_pred)

print('Confusion matrix = ')
print(MatrixC)
print('Testing Accuracy = {}'.format(ScoreA))
print('classification report = ')
print(ReportC)
```

## **Train Dataset dengan Algoritma NaiveBayes**

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
MatrixC = confusion_matrix(y_test, y_pred)
ReportC = classification_report(y_test,y_pred)
ScoreA = accuracy_score(y_test,y_pred)

print('Confusion matrix = ')
print(MatrixC)
print('Testing Accuracy = {}'.format(ScoreA))
print('classification report = ')
print(ReportC)
```

# **Train Dataset dengan SVM**

```python
classifier = SVC(kernel='linear',random_state=0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
ScoreA = accuracy_score(y_test,y_pred)
ReportC = classification_report(y_test,y_pred)
MatrixC = confusion_matrix(y_test, y_pred)
print('Confusion matrix = ')
print(MatrixC)
print('Testing Accuracy = {}'.format(ScoreA))
print('classification report = ')
print(ReportC)
```
