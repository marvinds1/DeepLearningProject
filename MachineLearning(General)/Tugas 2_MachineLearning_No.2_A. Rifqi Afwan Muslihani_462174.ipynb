Import dan ambil semua library/alat yang kita akan pakai

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
```

Ambil dataset yang ditelah disimpan di dalam drive

```python
habermandata = pd.read_csv('/content/gdrive/MyDrive/ColabNotebooks/haberman.data',header=None)
habermandata.head()
```

Split the dataset dengan tes_size = 0.2

```python
X = habermandata.iloc[:, [0, 1, 2]].values
y = habermandata.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Scaling data

```python
StandarS = StandardScaler()
X_train = StandarS.fit_transform(X_train)
X_test = StandarS.transform(X_test)
```

Dengan Algoritma KNN untuk mencari parameter terbaik

```python
knn = KNeighborsClassifier(n_neighbors=5)
k_range = range(1,31)
GridP = dict(n_neighbors=k_range)
Grid = GridSearchCV(knn, param_grid=GridP, cv=5, scoring='roc_auc')
print(Grid)
```

Menampilkan hasil parameter terbaik

```python
Grid.fit(X_train,y_train)
print(Grid.best_score_)
print(Grid.best_params_)
print(Grid.best_estimator_)
```

Dengan SVC Classifier untuk mencari parameter terbaik

```python
classifier = SVC(kernel='linear',random_state=0)
GridP = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}
 
GridSVC = GridSearchCV(classifier, GridP, cv=5, scoring='roc_auc')
GridSVC.fit(X_train, y_train)

print(GridSVC.best_score_)
print(GridSVC.best_params_)
```

Dengan Bernoulli Naive Bayes untuk mencari parameter terbaik

```python
GridP = {'alpha': [0.01, 0.1, 0.5, 1.0, 10.0],
         }

GridNB = GridSearchCV(BernoulliNB(), GridP, cv=5, scoring='roc_auc')
GridNB.fit(X,y)
print(GridNB.best_score_)
print(GridNB.best_params_)
```

Didapatkan bahwa skor tertinggi dari semua yang classifier adalah Algoritma KNN

```python
y_pred = Grid.predict(X_test)
```

```python
MatrixC = confusion_matrix(y_test, y_pred)
ScoreA = accuracy_score(y_test,y_pred)
ReportC = classification_report(y_test,y_pred)

print(MatrixC)
print(ScoreA)
print(ReportC)
```
